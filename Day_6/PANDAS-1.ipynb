```python
import numpy as np
import pandas as pd

# Create DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Ella'],
    'Age': [25, 30, 35, 40, np.nan],
    'Salary': [50000, 54000, 50000, 48000, 47000],
    'Department': ['HR', 'Finance', 'HR', 'Tech', 'Tech']
}

df = pd.DataFrame(data)

print("========== ORIGINAL DATAFRAME ==========")
print(df)

# Find Median
median_age = df['Age'].median()
print("\nMedian Age is:", median_age)

# Fill null values with median
df['Age'] = df['Age'].fillna(median_age)

print("\n========== UPDATED DATAFRAME ==========")
print(df)
```

```python
# Create DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Ella'],
    'Age': [25, 30, np.nan, 40, 28],
    'Salary': [50000, 54000, 55000, np.nan, 47000],
    'Department': ['HR', 'Finance', 'HR', 'Tech', np.nan]
}

df = pd.DataFrame(data)
print("========== ORIGINAL DATAFRAME ==========")
print(df)

# Drop rows with null values
df.dropna(inplace=True) #implace=True modifies the original DataFrame
print("\n========== DATAFRAME AFTER DROPPING NULL VALUES ==========")
print(df)
```

```python
# EMPLOYEE DATAFRAME
employee_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'Finance', 'HR', 'Tech'],
    'Salary': [50000, 54000, 55000, 53000]
}
df_employees = pd.DataFrame(employee_data)
print("========== EMPLOYEE DATAFRAME ==========")
print(df_employees)

# DEPARTMENT DATAFRAME
department_data = {
    'Department': ['HR', 'Finance', 'Tech'],
    'Budget': [100000, 150000, 200000]
}

df_departments = pd.DataFrame(department_data)
print("\n========== DEPARTMENT DATAFRAME ==========")
print(df_departments)

merged_df = pd.merge(df_employees,df_departments,on='Department') # Merge karte time wo common column dikha rha hai 
print("\n========== MERGED DATAFRAME ==========")
print(merged_df)
```

```python
# CREATE STUDENT DATAFRAME
data = {
    'Student_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Marks': [45, 78, 90, 32, 67, 88, 49, 95]
}
df = pd.DataFrame(data)
print("========== ORIGINAL DATAFRAME ==========")
print(df)

# APPLY FUNCTION USING LAMBDA
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')
print("\n========== UPDATED DATAFRAME ==========")
print(df)
```

```python

data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106],
    'Salary': [25000, 40000, 55000, 30000, 70000, 45000],
    'Seniority': ['Junior', 'Senior', 'Senior', 'Junior', 'Senior', 'Junior']
}

df = pd.DataFrame(data)

print("========== ORIGINAL DATAFRAME ==========")
print(df)

df['Updated_Salary'] = [x + 5000 if y == 'Senior' else x for x, y in zip(df['Salary'], df['Seniority'])] # zip one type of replace function hai jisme hum do list ko ek sath iterate kar sakte hai
print("\n========== UPDATED DATAFRAME ==========")
print(df)
```

```python
import pandas as pd

# ============================================
# CREATE MOVIE DATAFRAME
# ============================================

data = {
    'Movie': ['KGF', 'Pushpa', 'RRR', 'Jawan', 'Animal', 'Leo'],
    'Collection': [250, 120, 300, 180, 90, 220]
}
df = pd.DataFrame(data)

print("========== ORIGINAL DATAFRAME ==========")
print(df)

# USE qcut() FUNCTION
# Divide movies into categories based on collection

df['Movie_Status'] = pd.qcut(df['Collection'],q=3,labels=['Average', 'Hit', 'Blockbuster']) # qcut function is used to divide the data into equal quantiles.
print("\n========== DATAFRAME AFTER pd.qcut() ==========")
print(df)
```

```python
import pandas as pd
import io

# ============================================
# ORIGINAL DATAFRAME
# ============================================

employee_data = {
    'Employee_ID': [101, 102, 103],
    'Name': ['Amit', 'Riya', 'Karan'],
    'Department': ['HR', 'Finance', 'Tech'],
    'Salary': [50000, 65000, 72000]
}
df = pd.DataFrame(employee_data)
print("========== ORIGINAL DATAFRAME ==========")
print(df)


json_data = '''
[
    {
        "Employee_ID": 104,
        "Name": "Neha",
        "Department": "Marketing",
        "Salary": 48000
    },
    {
        "Employee_ID": 105,
        "Name": "Raj",
        "Department": "Sales",
        "Salary": 55000
    }
]
'''

# ============================================
# CONVERT JSON TO DATAFRAME
# ============================================

df_from_json = pd.read_json(io.StringIO(json_data))
print("\n========== DATAFRAME FROM JSON ==========")
print(df_from_json)

df_combined = pd.concat([df, df_from_json],ignore_index=False) # ignore_index=True is used to reset the index of the combined dataframe
print("\n========== COMBINED DATAFRAME ==========")
print(df_combined)
```

```python
import pandas as pd
data = {
    'Store': ['A', 'B', 'A', 'C', 'B', 'C', 'A'],
    'Sales': [12000, 18000, 15000, 20000, 17000, 22000, 14000],
    'Orders': [120, 150, 130, 180, 160, 210, 125]
}

df = pd.DataFrame(data)

print("========== ORIGINAL DATAFRAME ==========")
print(df)

grouped_df = df.groupby('Store').agg({
    'Sales': ['mean', 'min', 'max', 'sum'],
    'Orders': ['mean', 'min', 'max', 'sum']
}) # agg() function is used to apply multiple aggregation functions to the grouped data. and groupby() function is used to group the data based on a specific column.
print("\n========== GROUPED DATAFRAME ==========")
print(grouped_df)
```
