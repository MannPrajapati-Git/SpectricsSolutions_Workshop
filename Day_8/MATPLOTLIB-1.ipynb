```python
# 1. Plot a simple line graph of y = x² for x from 0 to 10

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
# 11. Plot multiple subplots in one figure using plt.subplot() or plt.subplots() # Using plt.subplots() (preferred):

fig, axs = plt.subplots(2, 2)  # 2x2 grid of subplots

x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, x)
axs[0, 1].plot(x, x**2)
axs[1, 0].plot(x, np.sin(x))
axs[1, 1].plot(x, np.log1p(x))

plt.tight_layout()
plt.show()
```

```python
# 12. Set custom tick marks and labels on axes

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xticks([0, 2.5, 5, 7.5, 10], ['Zero', 'Two.Five', 'Five', 'Seven.Five', 'Ten'])
plt.yticks([-1, 0, 1], ['Low', 'Zero', 'High'])
plt.show()
```

```python
# 13. Annotate specific points in a plot with plt.annotate()
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(2, 1.2),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

```python
#  14. Plot a filled area under a curve with fill_between()

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y, color='skyblue', alpha=0.9)
plt.show()
```

```python
# 15. Create a stacked bar chart

labels = ['A', 'B', 'C']
men = [20, 35, 30]
women = [25, 32, 34]

x = np.arange(len(labels))

plt.bar(x, men, label='Men')
plt.bar(x, women, bottom=men, label='Women')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

```python
# 16. Customize the figure size and DPI

plt.figure(figsize=(8, 4), dpi=100)
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.plot(x, y)
plt.title("Cosine Wave")
plt.show()
```

```python
# 17. Change font size, type, and color in titles and labels

plt.plot(x, y)
plt.title("Styled Title", fontsize=16, fontname='Comic Sans MS', color='purple')
plt.xlabel("X Axis", fontsize=12, color='green')
plt.ylabel("Y Axis", fontsize=12, color='blue')
plt.show()
```

```python
# 18. Use different color maps in scatter or contour plots

x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()  # Shows color scale
plt.show()
```

```python
#  19. Add text labels to bars in a bar chart

labels = ['Python', 'Java', 'C++']
values = [90, 60, 80]

bars = plt.bar(labels, values)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height - 5, str(height),
             ha='center', va='bottom', color='white')

plt.ylim(0, 100)
plt.show()
```

```python
# 20. Create a plot with twin y-axes (twinx())

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x / 10)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-', label='Sine')
ax2.plot(x, y2, 'b-', label='Exponential')

ax1.set_xlabel('X data')
ax1.set_ylabel('Sine', color='g')
ax2.set_ylabel('Exponential', color='b')

plt.title("Twin Y-Axes")
plt.show()
```
