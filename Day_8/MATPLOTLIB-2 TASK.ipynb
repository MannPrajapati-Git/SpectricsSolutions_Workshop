What is Matplotlib?
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

```python
import matplotlib

print(matplotlib.__version__)


```

Pyplot

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 54]

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', linestyle='--', color='green')
plt.title("Line with Markers")
plt.show()
```

```python
categories = ['Apple', 'Banana', 'Mango']
values = [40, 70, 30]

plt.bar(categories, values, color='orange')
plt.title("Fruit Sales")
plt.show()
```

```python
plt.barh(categories, values, color='purple')
plt.title("Horizontal Fruit Sales")
plt.show()
```

```python
labels = ['Python', 'C++', 'Java']
sizes = [45, 30, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
plt.axis('equal')
plt.show()
```

```python
x = [5, 6, 7, 8, 9]
y = [1, 3, 2, 5, 4]

plt.scatter(x, y, color='red')
plt.title("Simple Scatter Plot")
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

plt.plot(x, y1, label="Even", color='blue')
plt.bar(x, y2, label="Odd", color='green')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Left Plot")

plt.subplot(1, 2, 2)
plt.bar(['A', 'B', 'C'], [3, 5, 7])
plt.title("Right Plot")

plt.tight_layout()
plt.show()
```

```python

x = [1, 2, 3, 4, 5, 1]
y = [3, 5, 2, 8, 7, 8]

plt.plot(x, y)
plt.title("Styled Line Plot")
plt.show()
```

```python
x = [1, 2, 3]
y = [10, 20, 30]

plt.plot(x, y)
plt.title("Saved Plot Example")
plt.savefig("my_plot.png")  # Saves the figure as an image
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```
